```python
import os
import requests
import json
import time

folder = "C:\\Users\\03520806\\Documents\\Valida_API_teste"

def consulta_cnpj(cnpj):

    url = f"https://open.cnpja.com/office/{cnpj}"

    try:
        response = requests.request("GET", url, timeout=10)
        response.raise_for_status()
        resposta = json.loads(response.text)

        if "company" in resposta and "simples" in resposta["company"] and "optant" in resposta["company"]["simples"]:
            return resposta["company"]["simples"]["optant"]
        else:
            return "Informação indisponível"
        
    except requests.exceptions.RequestException as e:
        return f"Erro na requisição: {e}"
    
    except json.JSONDecodeError:
        return "Erro ao decodificar a resposta JSON"
    
def processar_arquivos(pasta):

    resultados = []
    contador = 0
    
    for arquivo in os.listdir(pasta):
        if arquivo.endswith('.txt'):
            with open(os.path.join(pasta, arquivo), 'r', encoding='latin-1') as f:
                for linha in f:
                    cnpj = linha.strip()  
                    status = consulta_cnpj(cnpj)
                    resultados.append(f"{cnpj}:{status}")

                    contador += 1

                    if contador % 5 == 0:
                        time.sleep(61)
    
    return resultados

def salvar_resultados(resultados, arquivo_saida):

    with open(arquivo_saida, 'w', encoding='latin-1') as f:
        for linha in resultados:
            f.write(f"{linha}\n")

pasta = folder
resultado = processar_arquivos(pasta)
arquivo_saida = "resultados.txt"


salvar_resultados(resultado, arquivo_saida)
    
```
